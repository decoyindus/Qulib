```python
import flask
from flask import request
import numpy as np
# importing Qiskit
from qiskit import BasicAer
from qiskit import QuantumCircuit, assemble, execute,ClassicalRegister
# import basic plot tools
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
from random import choice
import io
import json
import base64
from qiskit.circuit import qpy_serialization
from qiskit.aqua.components.oracles import TruthTableOracle
import operator
import requests
```

```python
URL = 'http://127.0.0.1:5000/demo/get_oracle'
p = {
    'n':'3',
    'case':'b',
    'measure':'True'
}
response = requests.get(URL,params=p)
```

```python
circuit_json = response.json()['oracle']
qpy_file = io.BytesIO(base64.b64decode(circuit_json))
circuit = qpy_serialization.load(qpy_file)[0]
```

```python
circuit.draw(output='mpl')
```

```python
URL2 = 'http://127.0.0.1:5000/demo/get_type'
buf = io.BytesIO()
qpy_serialization.dump(circuit, buf)
cstr= base64.b64encode(buf.getvalue()).decode('utf8')
json_str = {
        'circuit': cstr
    }

response = requests.get(URL2,params=json_str)
response.json()
```
