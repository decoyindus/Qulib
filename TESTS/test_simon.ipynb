```python
import base64
import json
import requests
import io
from qiskit.circuit import qpy_serialization
```

```python
URL = 'https://qulib-oracle.herokuapp.com/demo/get_simon_oracle'
p = {'key':'110'}
response = requests.get(URL, params=p)
oracle_json = response.json()['oracle']
qpy_file = io.BytesIO(base64.b64decode(oracle_json))
oracle = qpy_serialization.load(qpy_file)[0]
oracle.draw(output = 'mpl')
```

```python
URL_KEY = 'https://qulib-oracle.herokuapp.com/demo/get_simon_key'
prm = {'circuit':oracle_json}
response = requests.get(URL_KEY,params=prm)
response.json()['key']
```
